```
import sklearn
from sklearn.model_selection import train_test_split
from sklearn import linear_model
from sklearn.svm import SVR
from sklearn.preprocessing import StandardScaler
from sklearn import metrics
```

```
df_A = df.copy()
df_A.head(10)
```

```
X = np.array(df_A['sqft_living']).reshape(-1,1)
y = np.array(df_A['price']).reshape(-1,1)
```

```
##12
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.2)
X_train
```

```
##13
sc_X = StandardScaler()
sc_y = StandardScaler()
X_tran = sc_X.fit_transform(X_train)
y_tran = sc_y.fit_transform(y_train)
X_test_tran = sc_X.fit_transform(X_test)
y_test_tran = sc_y.fit_transform(y_test)
```

```
regressor = SVR(kernel='linear')
regressor_rbf = SVR(kernel='rbf')
```

**Linear**

```
##14
regressor.fit(X_tran, y_tran)
predicted_lin = sc_y.inverse_transform(regressor.predict(X_test_tran))
predicted_lin
```

**RBF**

```
regressor_rbf.fit(X_tran, y_tran)
predicted_rbf = sc_y.inverse_transform(regressor_rbf.predict(X_test_tran))
predicted_rbf
```

**Linear**

```
##15
print('MAE :', metrics.mean_absolute_error(y_test, predicted_lin))
print('MSE :', metrics.mean_squared_error(y_test, predicted_lin))
print('RMSE :', np.sqrt(metrics.mean_squared_error(y_test, predicted_lin)))
```

**RBF**

```
print('MAE :', metrics.mean_absolute_error(y_test, predicted_rbf))
print('MSE :', metrics.mean_squared_error(y_test, predicted_rbf))
print('RMSE :', np.sqrt(metrics.mean_squared_error(y_test, predicted_rbf)))
```

```
X = np.array(df['sqft_living']).reshape(-1, 1)
y = np.array(df['price']).reshape(-1, 1)
```

```
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.2, random_state= 100)
```

```
sc_X = StandardScaler()
sc_y = StandardScaler()
X_train = sc_X.fit_transform(X_train)
y_train = sc_y.fit_transform(y_train)
```

```
svr_reg = SVR(kernel= 'rbf')
svr_reg.fit(X_train, y_train)
```

```
X_test = sc_X.fit_transform(X_test)
y_test = sc_y.fit_transform(y_test)
```

```
pred_lin = sc_y.inverse_transform(svr_reg.predict(X_test))
pred_lin
```

```
print('MAE :', metrics.mean_absolute_error(y_test, pred_lin))
print('MSE :', metrics.mean_squared_error(y_test, pred_lin))
print('RMSE :', np.sqrt(metrics.mean_squared_error(y_test, pred_lin)))
```
