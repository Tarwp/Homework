```
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix, accuracy_score, f1_score, precision_score, recall_score
```

```
X = df.drop(['name'], axis= 1)
y = df['name']
```

```
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.2, random_state= 100)
```

```
X_std = StandardScaler()
X_nor = MinMaxScaler()
```

```
knn = KNeighborsClassifier()
```

```
##16
knn.fit(X_train, y_train)
predicted = knn.predict(X_test)
```

**Standardization**

```
X_std_train = X_std.fit_transform(X_train)
X_std_test = X_std.fit_transform(X_test)
```

```
knn.fit(X_std_train, y_train)
predicted_A = knn.predict(X_std_test)
```

**Normalization**

```
X_nor_train = X_nor.fit_transform(X_train)
X_nor_test = X_nor.fit_transform(X_test)
```

```
knn.fit(X_nor_train, y_train)
predicted_B = knn.predict(X_nor_test)
```

**Without Standardization & Normalization**

```
##15
print('Accuracy :', accuracy_score(y_test, predicted))
print('F1 Score :', f1_score(y_test, predicted, average='micro'))
print('Recall :', recall_score(y_test, predicted, average='micro'))
print('precision :', precision_score(y_test, predicted, average='micro'))
```

```
confusion_matrix(y_test, predicted)
```

**Standardization**

```
print('Accuracy :', accuracy_score(y_test, predicted_A))
print('F1 Score :', f1_score(y_test, predicted_A, average='micro'))
print('Recall :', recall_score(y_test, predicted_A, average='micro'))
print('precision :', precision_score(y_test, predicted_A, average='micro'))
```

```
confusion_matrix(y_test, predicted_A)
```

**Normalization (The Best Choice)**

```
print('Accuracy :', accuracy_score(y_test, predicted_B))
print('F1 Score :', f1_score(y_test, predicted_B, average='micro'))
print('Recall :', recall_score(y_test, predicted_B, average='micro'))
print('precision :', precision_score(y_test, predicted_B, average='micro'))
```

```
confusion_matrix(y_test, predicted_B)
```

```
##17
accuracy_list = []
for i in range(1, 50) :
  knn = KNeighborsClassifier(n_neighbors= i)
  knn.fit(X_nor_train, y_train)
  predicted_i = knn.predict(X_nor_test)
  accuracy_list.append(accuracy_score(y_test, predicted_i))
```

```
accuracy_list
```

```
plt.figure(figsize=(10, 8))
plt.plot(range(1, 50), accuracy_list, color= 'black', linestyle= 'dashed', marker= 'o')
```
