```
cancer = {'M' : 1, 'B' : 0}
df['diagnosis_result'] = df['diagnosis_result'].map(cancer)
```

```
X =df.drop(['id', 'diagnosis_result'], axis= 1)
y = df['diagnosis_result']
```

```
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import confusion_matrix, accuracy_score, f1_score, precision_score, recall_score
```

```
svc = SVC()
X_std = StandardScaler()
X_nom = MinMaxScaler()
```

```
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.3, random_state= 100)
```

**Without Standization & Normalization**

```
##15 & 16
svc.fit(X_train, y_train)
predicted = svc.predict(X_test)
```

```
confusion_matrix(y_test, predicted)
```

```
print(accuracy_score(y_test, predicted))
print(f1_score(y_test, predicted))
print(precision_score(y_test, predicted))
print(recall_score(y_test, predicted))
```

**Standradization**

```
X_train_std = X_std.fit_transform(X_train)
X_test_std = X_std.fit_transform(X_test)
svc.fit(X_train_std, y_train)
predicted_std = svc.predict(X_test_std)
```

```
confusion_matrix(y_test, predicted_std)
```

```
print(accuracy_score(y_test, predicted_std))
print(f1_score(y_test, predicted_std))
print(precision_score(y_test, predicted_std))
print(recall_score(y_test, predicted_std))
```

**Normalization**

```
X_train_nom = X_nom.fit_transform(X_train)
X_test_nom = X_nom.fit_transform(X_test)
svc.fit(X_train_nom, y_train)
predicted_nom = svc.predict(X_test_nom)
```

```
confusion_matrix(y_test, predicted_nom)
```

```
print(accuracy_score(y_test, predicted_nom))
print(f1_score(y_test, predicted_nom))
print(precision_score(y_test, predicted_nom))
print(recall_score(y_test, predicted_nom))
```

```
##17
param_combination = {'C' : [0.01, 0.1, 1, 10, 100, 100], 'gamma' : [0.00001, 0.0001, 0.001, 0.01, 0.1, 0.100]}
grid_search = GridSearchCV(SVC(), param_combination, verbose= 1)
grid_search.fit(X_train, y_train)
grid_search.best_params_
```

```
grid_predicted = grid_search.predict(X_test)
```

```
confusion_matrix(y_test, grid_predicted)
```

```
print(accuracy_score(y_test, grid_predicted))
print(f1_score(y_test, grid_predicted))
print(precision_score(y_test, grid_predicted))
print(recall_score(y_test, grid_predicted))
```
