```
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score, confusion_matrix
```

```
##14
df = df.replace({'\t?' : np.nan})
df['age'] = df['age'].fillna(df['age'].mean())
df['bp'] = df['bp'].fillna(df['bp'].mean())
df['sg'] = df['sg'].fillna(df['sg'].mean())
df['al'] = df['al'].fillna(df['al'].mean())
df['al'] = df['al'].fillna(df['al'].mean())
df['su'] = df['su'].fillna(df['su'].mean())
df['classification'] = df['classification'].apply(lambda x: x.split('\t')[0])
df['bgr'] = df['bgr'].fillna(pd.Series(np.where(df['classification'] == 'notcdk', 107.72, 175.41)))
df['bu'] = df['bu'].fillna(pd.Series(np.where(df['classification'] == 'notcdk', 32.80, 72.39)))
df['sc'] = df['sc'].fillna(df['sc'].mean())
df['sod'] = df['sod'].fillna(df['sod'].mean())
df['pot'] = df['pot'].fillna(df['pot'].mean())
df['hemo'] = df['hemo'].fillna(df['hemo'].mean())
```

```
df['rbc'] = df['rbc'].fillna(pd.Series(np.where(df['classification'] == 'notcdk', 'abnormal', 'normal')))
df['pc'] = df['pc'].fillna('normal')
df['pcc'] = df['pcc'].fillna('notpresent')
df['ba'] = df['ba'].fillna('notpresent')
df['pcv'] = df['pcv'].astype('float64')
df['pcv'] = df['pcv'].fillna(df['pcv'].mean())
df['wc'] = df['wc'].astype('float64')
df['wc'] = df['wc'].fillna(df['wc'].mean())
df['rc'] = df['rc'].astype('float64')
df['rc'] = df['rc'].fillna(df['rc'].mean())
df['htn'] = df['htn'].fillna('no')
df['dm'] = df['dm'].fillna('no')
df['cad'] = df['cad'].fillna('no')
df['appet'] = df['appet'].fillna('good')
df['pe'] = df['pe'].fillna('no')
df['ane'] = df['ane'].fillna('no')
```

```
df['dm'] = df['dm'].replace({'\tno' : 'no', '\tyes' : 'yes', ' yes' : 'yes'})
df['cad'] = df['cad'].replace({'\tno' : 'no'})
df[['htn', 'dm', 'cad', 'pe', 'ane']] = df[['htn', 'dm', 'cad', 'pe', 'ane']].replace({'yes' : 1, 'no' : 0})
df[['rbc', 'pc']] = df[['rbc', 'pc']].replace({'normal' : 1, 'abnormal' : 0})
df[['pcc', 'ba']] = df[['pcc', 'ba']].replace({'present' : 1, 'notpresent' : 0})
df['appet'] = df['appet'].replace({'good' : 1, 'poor' : 0})
df['classification'] = df['classification'].replace({'ckd' : 1, 'notckd' : 0})
```

```
##12
X = df[['al', 'su', 'bgr', 'bu', 'htn', 'dm', 'pe', 'ane']]
y = df[['classification']]
X1 = df.drop(['id', 'classification'], axis=1 )
y1 = df[['classification']]
X_trian, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.3, random_state = 100)
X1_train, X1_test, y1_train, y1_test = train_test_split(X1, y1, test_size= 0.3, random_state = 100)
```

```
log_reg.fit(X_trian, y_train)
predicted = log_reg.predict(X_test)
```
