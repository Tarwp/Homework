```
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
%matplotlib inline
```

```
##1
x = np.arange(1,21)
y = x**2
plt.title('XY')
plt.xlabel('X label')
plt.ylabel('Y label')
plt.plot(x, y)
```

```
##2
x = np.arange(0, 99)
y = np.sin(x)
plt.title('XY')
plt.xlabel('X label')
plt.ylabel('Y label')
plt.plot(x, y, color= 'y')
```

```
##3
y = np.arange(0, 20)
x = y**2
plt.title('XY')
plt.xlabel('X label')
plt.ylabel('Y label')
plt.plot(x, y)
```

```
##4
x = np.arange(0, 99)
y = np.cos(x)
plt.title('XY')
plt.xlabel('X label')
plt.ylabel('Y label')
plt.plot(x, y, ls= '--', color= 'r')
```

```
##5
x = np.arange(0, 25)
x = np.arange(0, 50)
plt.plot(x, x**2, label= 'y1', marker= '.')
plt.plot(x, x**3, label= 'y2', marker= '^')
plt.title('y1 & y2')
plt.xlabel('X label')
plt.ylabel('Y label')
plt.legend()
```

```
##6
x = np.arange(0, 50)
y = np.arange(0, 100)

plt.plot(x, x**2, label= 'y1', marker= '.', color= 'r')
plt.plot(y**2, y, label= 'y2', marker= '^', color= 'b')
plt.title('y1 & y2')
plt.xlabel('X label')
plt.ylabel('Y label')
plt.grid()
```

```
##7
y = np.arange(0,50)
x = y**2+4*y
plt.plot(x, y, label= 'y1', marker= '.', color= '#FFF56E')

plt.title('y1 & y2')
plt.xlabel('X label')
plt.ylabel('Y label')
```

```
##8
x = np.arange(0, 25)
y = np.arange(0, 100)
fig = plt.figure()
ax1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
ax2 = fig.add_axes([0.4, 0.15, 0.4, 0.4])
ax1.plot(y**3, y)
ax2.plot(x, x**2 + 2)
```

```
df = pd.read_csv('titanic_train.csv')
df.sample()
```

```
age_avg = df['Age'].mean()
df['Age'].fillna(age_avg, inplace = True)
```

```
##9
p_class = df['Pclass'].value_counts()
p_class.plot.pie(subplots = True, figsize = (6, 6), autopct= '%.2f%%')
```

```
##10
em = df['Embarked'].value_counts()
em
explode = [0, 0, 0.1]
em.plot.pie(subplots = True, figsize = (6, 6), autopct= '%.2f%%', explode= explode)
```

```
##11
df.head()
plt.hist(df['Fare'], bins= 50)
```

```
##12
plt.scatter(x= df['Age'], y= df['Fare'])
```

```
##13
avg_f = df.groupby('Pclass').mean()['Fare']
avg_f.plot(kind= 'bar')
```

```
##14
df.groupby('Age').mean()['Fare'].plot(kind= 'bar', figsize = (15, 5))
```

```
##15
df['Surname'] = df['Name'].apply(lambda x: x.split(',')[0])
df_d = df['Surname'].value_counts()
df_d[df_d > 1].plot(kind= 'bar', figsize = (20, 5))
```

```
##16
sur_class = df.groupby('Pclass').count()['Survived']
s_fm = df[(df['Pclass']) & (df['Survived'] == 1)]['Sex'].value_counts()

```

```
##16 Non-sur
ns_fm = df[(df['Pclass']) & (df['Survived'] == 0)]['Sex'].value_counts()
ns_fm
```

```
##16
plt.subplot(321, title= 'Survived Family/Group/Couple Tourist')
df[(df['SibSp'] > 0)]['Survived'].value_counts().plot(kind= 'bar')
plt.subplot(323, title= 'Survived Alone Tourist')
df[(df['SibSp'] == 0)]['Survived'].value_counts().plot(kind= 'bar')
plt.subplot(325, title= 'Survived Rate by Embarked')
df[df['Survived'] == 1]['Embarked'].value_counts().plot(kind= 'bar')
plt.subplot(222, title= 'Male & Femal Servival rates', xlabel= 'Gender')
s_fm.plot(kind= 'bar')
plt.xlabel('Gender')
plt.subplot(224, title= 'Male & Femal Non-Servival rates')
ns_fm.plot(kind= 'bar')
plt.xlabel('Gender')
plt.tight_layout()
```
