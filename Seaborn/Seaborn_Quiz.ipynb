```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```
##1
com = ['A', 'B', 'C', 'D']
years = [2014, 2015, 2016, 2017]
values = np.random.uniform(50,501,(4,4))
df_1 = pd.DataFrame(values, com, years)
df_1['Avg'] = df_1.sum(axis=1)/4
df_1
```

```
##2
sns.barplot(x= df_1.index, y= df_1['Avg'])
```

```
sns.barplot(x= df_1['Avg'], y= df_1.index)
```

```
##3
sns.scatterplot(x= df_1.index, y= df_1[2017])
```

```
sns.scatterplot(x= df_1[2017], y= df_1.index)
```
